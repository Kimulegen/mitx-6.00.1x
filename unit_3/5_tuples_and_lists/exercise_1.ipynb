When a tuple has only one element, you must specify it as follows: (elt,). Here is an example shell session that illustrates the difference:

```

>>> type(tup1)
<type 'int'>
>>> tup2 = (5,)
>>> print(tup2)
(5,)
>>> type(tup2)
<type 'tuple'>
```
